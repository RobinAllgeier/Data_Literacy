```python
import pandas as pd
import matplotlib.pyplot as plt

fileName19 = "Stadtbücherei Tübingen - 2019.csv"
fileName20 = "Stadtbücherei Tübingen - 2020.csv"
fileName21 = "Stadtbücherei Tübingen - 2021.csv"
fileName22 = "Stadtbücherei Tübingen - 2022.csv"
fileName23 = "Stadtbücherei Tübingen - 2023.csv"
fileName24 = "Stadtbücherei Tübingen - 2024.csv"
fileName25 = "Stadtbücherei Tübingen - 01.01. bis  06.11. - 2025.csv"
dataDirectoryPath  = "../Jaehrliche_csv/"
ausleihdatumColumn = 'Ausleihdatum/Uhrzeit'

data19 = pd.read_csv(dataDirectoryPath + fileName19, sep=";")
data20 = pd.read_csv(dataDirectoryPath + fileName20, sep=";")
data21 = pd.read_csv(dataDirectoryPath + fileName21, sep=";")
data22 = pd.read_csv(dataDirectoryPath + fileName22, sep=";")
data23 = pd.read_csv(dataDirectoryPath + fileName23, sep=";")
data24 = pd.read_csv(dataDirectoryPath + fileName24, sep=";")
data25 = pd.read_csv(dataDirectoryPath + fileName25, sep=";")

def getLoansPerDay(data):
    data[ausleihdatumColumn] = pd.to_datetime(data[ausleihdatumColumn])
    return data.groupby(data[ausleihdatumColumn].dt.date).size()

def plotYearData(loansPerDay, title): 
    plt.figure(figsize=(12,6))
    loansPerDay.plot(kind='line')
    plt.xlabel("Date")
    plt.ylabel("Number of Loans")
    plt.title(title)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.xlim(loansPerDay.index.min(), loansPerDay.index.max())
    plt.ylim(bottom=0)

    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b'))  # Jan, Feb, ...
    plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MonthLocator())
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

```

```python
plotYearData(getLoansPerDay(data19), "Loans per Day 2019")
```

```python
plotYearData(getLoansPerDay(data20), "Loans per Day 2020")
```

```python
plotYearData(getLoansPerDay(data21), "Loans per Day 2021")
```

```python
plotYearData(getLoansPerDay(data22), "Loans per Day 2022")
```

```python
plotYearData(getLoansPerDay(data23), "Loans per Day 2023")
```
